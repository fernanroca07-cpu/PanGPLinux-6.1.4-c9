```python
import numpy as np

# number of terms
NTERM = 10

# value of each term (single precision)
VALUE = np.float32(0.1)

sum_val = np.float32(0.0)

# for all terms
for index in range(1, NTERM+1):
    sum_val += VALUE 

# divide by total number of terms
sum_val /= np.float32(NTERM) 

# calcualte relative error
error = (sum_val - VALUE) / VALUE 

print(f"Sum={sum_val:g} Relative error = {error:g}")
```
