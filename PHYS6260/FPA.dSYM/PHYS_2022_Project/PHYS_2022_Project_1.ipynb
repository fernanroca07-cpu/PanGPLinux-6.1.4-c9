```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime

# 1. Create the dataset based on the bimonthly SOHO observations
data = {
    'Date': [
        '2016-02-15', '2016-04-15', '2016-06-15', '2016-08-15', '2016-10-15', '2016-12-15',
        '2017-02-15', '2017-04-15', '2017-06-15', '2017-08-15', '2017-10-15', '2017-12-15',
        '2018-02-15', '2018-04-15', '2018-06-15', '2018-08-15', '2018-10-15', '2018-12-15',
        '2019-02-15', '2019-04-15', '2019-06-15', '2019-08-15', '2019-10-15', '2019-12-15',
        '2020-02-15', '2020-04-15', '2020-06-15', '2020-08-15', '2020-10-15', '2020-12-15',
        '2021-02-15', '2021-04-15', '2021-06-15', '2021-08-15', '2021-10-15', '2021-12-15'
    ],
    'Wolf_Number': [
        31, 49, 14, 34, 13, 18, 12, 30, 0, 16, 14, 0,
        0, 0, 12, 0, 0, 0, 0, 15, 0, 0, 0, 0,
        0, 11, 0, 12, 13, 32, 11, 34, 16, 51, 35, 78
    ]
}

# 2. Convert to DataFrame and format dates
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# 3. Initialize the plot
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Wolf_Number'], marker='o', linestyle='-', color='crimson', linewidth=1.5, markersize=5, label='Calculated Wolf Number (R)')

# 4. Formatting for scientific report
plt.title('Calculated Wolf Sunspot Number Progression (2016–2021)', fontsize=14, fontweight='bold')
plt.xlabel('Observation Year', fontsize=12)
plt.ylabel('Wolf Number ($R = k(10g + s)$)', fontsize=12) # Based on formula [cite: 48]
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Highlight Solar Minimum (approx. late 2019)
plt.axvspan(datetime.datetime(2018, 6, 1), datetime.datetime(2020, 1, 1), color='gray', alpha=0.2, label='Solar Minimum Phase')

# 5. Show and save
plt.tight_layout()
plt.show()
```
