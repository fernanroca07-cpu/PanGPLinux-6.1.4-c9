```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

# order for interpolation=number of points+1
n = 9

i = np.arange(n+1)
x = 2.0 * i/(1.0 * n) - 1.0

# runge function
f = 1.0 / (1.0 + 25.0 * x * x)

# plot function samples
plt.plot(x, f, "ro")
plt.axis([-1.2, 1.2, -0.2, 1.2])

# calculate lagrange interpolation
poly = interpolate.lagrange(x, f)

# evaluate polynomial in a fine grid of points
xn = np.linspace(-1.0, 1.0, 128)
yn = poly(xn)

# plot interpolating polynomial
plt.plot(xn, yn, "k-") 

# evaluate function
fn = 1.0 / (1.0 + 25.0 * xn * xn)
plt.plot(xn, fn, "b--") 

plt.show()
```
